# This notebook is used to draw figures for the microbenchmarks.

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.size'] = 32

xs = ['0.01', '0.1',  '1', '10'] # scale factor

# checkpoint 0.1 done.

filtering = np.array([0,0,0,0])
projection = np.array([1.507, 6.554, 52.071, 278.925])
join = np.array([3.789, 24.114, 57.346, 102.335])
aggregation = np.array([6.36, 827.002, 27910, 253500])
union = np.array([30.922, 38.642, 41.089, 38.965])

colors = sns.color_palette("rocket", 5)

df = pd.DataFrame({
	'Scale Factor': xs,
	'$\mathtt{filter}$': filtering / 1000,
	'$\mathtt{project}$': projection/ 1000,
	'$\mathtt{join}$': join/ 1000,
	'$\mathtt{aggregate}$': aggregation/ 1000,
	'$\mathtt{union}$': union/ 1000
})

ax = df.set_index('Scale Factor').plot(kind='bar', stacked=True, color=colors, figsize=(10, 10))
plt.ticklabel_format(style='plain', useOffset=False, axis='y') ## No offset
plt.xticks(rotation=0) ## Rotate X-axis labels
ax.set_ylabel("Latency (ms)") ## Set Y-axis
ax.set_xlabel("Scale Factor") ## Set X-axis
ax.set_yscale('log') ## Set log scale
ax.legend()

plt.savefig('microbenchmark-policy-a.pdf', dpi=300, bbox_inches='tight') # no policy at all!
```

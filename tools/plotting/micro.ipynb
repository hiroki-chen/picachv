# This notebook is used to draw figures for the microbenchmarks.

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = 32

xs = ['0.01', '0.1', '1', '10'] # scale factor

filtering = np.array([0,0,0,0] )
projection = np.array([9, 35.6, 2114, 146340])
join = np.array([5.667, 36.51, 123.533, 534.703])
aggregation = np.array([2.198, 611.617, 7860, 19646])
union = np.array([269.199, 329.289, 399.741, 347.27])

colors = sns.color_palette("rocket", 5)

df = pd.DataFrame({
	'Scale Factor': xs,
	'Filtering': filtering / 1000,
	'Projection': projection/ 1000,
	'Join': join/ 1000,
	'Aggregation': aggregation/ 1000,
	'Union': union/ 1000
})

ax = df.set_index('Scale Factor').plot(kind='bar', stacked=True, color=colors, figsize=(10, 10))
plt.ticklabel_format(style='plain', useOffset=False, axis='y') ## No offset
plt.xticks(rotation=0) ## Rotate X-axis labels
ax.set_ylabel("Latency (ms)") ## Set Y-axis
ax.set_xlabel("Scale Factor") ## Set X-axis
ax.set_yscale('log') ## Set log scale
ax.legend()

plt.savefig('microbenchmark-policy-a.pdf', dpi=300, bbox_inches='tight') # no policy at all!
```

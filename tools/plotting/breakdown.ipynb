```python

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = 32
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

xs = ['0.01', '0.1', '1', '10'] # scale factor

policy_eval = np.array([0.076, 15.628, 167.338, 1005.667])
process = np.array([4.003, 17.953, 291.29, 1184.085])

# convert to percentage.
for i in range(len(policy_eval)):
	sum = policy_eval[i] + process[i]
	print(sum)
 
	policy_eval[i] = policy_eval[i] / sum
	process[i] = process[i] / sum


colors = sns.color_palette("rocket", 5)
df = pd.DataFrame({
	'Scale Factor': xs,
	'$\mathtt{policy\_eval}$': policy_eval,
	'$\mathtt{process}$': process,
})

print(df)


ax = df.set_index('Scale Factor').plot(kind='bar', stacked=True, color=colors, figsize=(10, 10))
plt.ticklabel_format(style='plain', useOffset=False, axis='y') ## No offset
plt.xticks(rotation=0) ## Rotate X-axis labels
ax.set_ylabel("Percentage") ## Set Y-axis
ax.set_xlabel("Scale Factor") ## Set X-axis

ax.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",)

plt.savefig('microbenchmark-project.pdf', dpi=300, bbox_inches='tight')
```

```python

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams['font.size'] = 32
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

xs = ['0.01', '0.1', '1', '10'] # scale factor

groupby = np.array([1.009, 6.946, 33.67, 168.074])
policy_eval_gb = np.array([0.050, 25.837, 215.593, 1369.87])
policy_eval_agg = np.array([2.564, 9.408, 179.963, 1887.962])
process = np.array([0.158, 12.919, 79.432, 284.28])

# convert to percentage.
for i in range(len(groupby)):
	sum = groupby[i] + policy_eval_gb[i] + policy_eval_agg[i] + process[i]
	groupby[i] = groupby[i] / sum
	policy_eval_gb[i] = policy_eval_gb[i] / sum
	policy_eval_agg[i] = policy_eval_agg[i] / sum
	process[i] = process[i] / sum 

colors = sns.color_palette("rocket", 5)
df = pd.DataFrame({
	'Scale Factor': xs,
	'$\mathtt{groupby}$': groupby,
	'$\mathtt{policy\_eval\_gb}$': policy_eval_gb,
	'$\mathtt{policy\_eval\_agg}$': policy_eval_agg,
	'$\mathtt{process}$': process,
})

ax = df.set_index('Scale Factor').plot(kind='bar', stacked=True, color=colors, figsize=(10, 10))
plt.ticklabel_format(style='plain', useOffset=False, axis='y') ## No offset
plt.xticks(rotation=0) ## Rotate X-axis labels
ax.set_ylabel("Percentage") ## Set Y-axis
ax.set_xlabel("Scale Factor") ## Set X-axis
ax.legend(bbox_to_anchor=(0, 1.02, 1, 0.2), loc="lower left",)


plt.savefig('microbenchmark-agg.pdf', dpi=300, bbox_inches='tight')
```
